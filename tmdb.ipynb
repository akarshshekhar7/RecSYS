```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from rake_nltk import Rake
import re
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.head()
```

```python
movies.rename(columns={'id':'movie_id'},inplace=True)

df = pd.merge(movies,credits,on='movie_id')

```

```python
df.head
```

```python
df.columns
```

```python
df.drop(columns=['budget','homepage','original_language','revenue', 'runtime','spoken_languages','status','title_x','title_y','popularity','release_date','vote_average','vote_count'],inplace=True)
```

```python
df.columns
```

```python
df.head(1)
```

```python

import json
def json_decode(data,key):
    result=[]
    data=json.loads(data)
    for item in data:
        result.append(item[key])
    return (result)

```

```python

df['genres']=df['genres'].apply(json_decode,key='name')

df['keywords']=df['keywords'].apply(json_decode,key='name')

df['production_companies']=df['production_companies'].apply(json_decode,key='name')

df['production_countries']=df['production_countries'].apply(json_decode,key='name')

df['cast']=df['cast'].apply(json_decode,key='name')

df['crew']=df['crew'].apply(json_decode,key='name')

```

```python
df.head(1)
```

```python


df.info()

```

```python


df['tagline'].fillna('',inplace=True)
df['overview'].fillna('',inplace=True)

```

```python
df.info()
```

```python
def remove_spaces(row):
    x=[]
    for i in row:
        i=i.replace(' ','')
        i=i.lower()
        x.append(i)
    return(x)
```

```python
df['genres'] = df['genres'].apply(remove_spaces).str.join(' ')

df['keywords'] = df['keywords'].apply(remove_spaces).str.join(' ')

df['production_countries'] = df['production_countries'].apply(remove_spaces).str.join(' ')

df['production_companies'] = df['production_companies'].apply(remove_spaces).str.join(' ')

df['cast'] = df['cast'].apply(remove_spaces).str.join(' ')

df['crew'] = df['crew'].apply(remove_spaces).str.join(' ')

df['overview'] = df['overview'].str.lower()

df['tagline'] = df['tagline'].str.lower()

df['original_title'] = df['original_title'].str.lower()

```

```python
df.head(1)
```

```python
from rake_nltk import Rake
r=Rake()
def extract_important_words(row):
    word = r.extract_keywords_from_text(row)
    word = r.get_ranked_phrases()
    return ' '.join(word)
```

```python


df['overview'] = df['overview'].apply(extract_important_words)

df['tagline'] = df['tagline'].apply(extract_important_words)

df['keywords'] = df['keywords'].apply(extract_important_words)

```

```python
df.head()
```

```python
df['combined_features'] = df['genres']+' '+df['keywords']+' '+df['overview']+' '+df['production_companies']+' '+df['production_countries']+' '+df['tagline']+' '+df['cast']+' '+df['crew']
```

```python
df.head(10)
```

```python
df.drop(columns=['movie_id','production_companies','production_countries','genres','keywords','overview','tagline','cast','crew'],inplace=True)
```

```python
df.head(1)
```

```python
def clear(data):
    result=[]
    result.append(re.sub("[^a-z1-9]","",data))
    return result
```

```python


df['title'] = df['original_title'].apply(clear).str.join('')

```

```python
df.to_csv('final_data.csv',index=False)
```
